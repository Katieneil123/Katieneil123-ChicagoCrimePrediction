```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import geopandas as gpd
from shapely.geometry import Point
import os
from datetime import datetime, timedelta
import warnings
import pickle
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
import warnings
warnings.filterwarnings('ignore')

# Data Path
DATA_DIR = "data/"
PROCESSED_DIR = "data/processed/"
MODELS_DIR = "models/"
```

## Some Useful Functions

```python
## Load Processed Data
def load_processed_data(filename):
    filepath = os.path.join(PROCESSED_DIR, filename)
    
    if os.path.exists(filepath):
        if filename.endswith('.csv'):
            return pd.read_csv(filepath, parse_dates=['Date'])
        elif filename.endswith('.parquet'):
            return pd.read_parquet(filepath)
        elif filename.endswith('.pkl'):
            with open(filepath, 'rb') as f:
                return pickle.load(f)
        else:
            print(f"Unsupported file format: {filename}")
            return None
    else:
        print(f"File not found: {filepath}")
        return None

#----- Phase 1: Basic Cleaning -----#

## Handle Missing Values
def handle_missing_values(df):

    result = df.copy()
    
    missing_before = result.isnull().sum()
    missing_pct_before = (missing_before / len(result)) * 100
    
    # Geospatial Data Imputation
    for coord_col in ['Latitude', 'Longitude', 'X Coordinate', 'Y Coordinate']:
        if coord_col in result.columns and result[coord_col].isnull().any():
            
            group_cols = ['Block', 'Location Description']
            median_coords = result.groupby(group_cols, observed=False)[coord_col].transform('median')
            result[coord_col] = result[coord_col].fillna(median_coords)            
            
            if result[coord_col].isnull().any():
                district_medians = result.groupby('District')[coord_col].transform('median')
                result[coord_col].fillna(district_medians, inplace=True)
            
            
            result[coord_col].fillna(result[coord_col].median(), inplace=True)
    
    # Categorical Data Imputation
    categorical_cols = result.select_dtypes(include=['object', 'category']).columns
    for col in categorical_cols:

        if col not in ['ID', 'Case Number'] and result[col].isnull().any():

            if pd.api.types.is_categorical_dtype(result[col]):
                most_common = result[col].value_counts().index[0]
                result[col].fillna(most_common, inplace=True)
            else:

                result[col].fillna("Unknown", inplace=True)
    
    numeric_cols = result.select_dtypes(include=['int', 'float']).columns
    for col in numeric_cols:
        if result[col].isnull().any():
            result[col].fillna(result[col].median(), inplace=True)
    
    # Analyze Missing Values
    missing_after = result.isnull().sum()
    missing_pct_after = (missing_after / len(result)) * 100
    
    # Generate Report
    report = {
        'missing_before': missing_before[missing_before > 0],
        'missing_pct_before': missing_pct_before[missing_pct_before > 0],
        'missing_after': missing_after[missing_after > 0],
        'missing_pct_after': missing_pct_after[missing_pct_after > 0],
        'filled_count': missing_before - missing_after
    }
    
    return result, report

# Standardize Date Format
def standardize_datetime(df):
    result = df.copy()
    
    if 'Date' in result.columns:
        if not pd.api.types.is_datetime64_dtype(result['Date']):
            result['Date'] = pd.to_datetime(result['Date'], errors='coerce')

        result['Year'] = result['Date'].dt.year
        result['Month'] = result['Date'].dt.month
        result['Day'] = result['Date'].dt.day
        result['DayOfWeek'] = result['Date'].dt.dayofweek
        result['Hour'] = result['Date'].dt.hour

        result['TimeOfDay'] = pd.cut(
            result['Hour'], 
            bins=[0, 6, 12, 18, 24], 
            labels=['Night', 'Morning', 'Afternoon', 'Evening'],
            include_lowest=True
        )
        
        result['IsWeekend'] = result['DayOfWeek'].isin([5, 6]).astype(int)
        # Season
        result['Season'] = pd.cut(
            result['Month'],
            bins=[0, 3, 6, 9, 12],
            labels=['Winter', 'Spring', 'Summer', 'Fall'],
            include_lowest=True
        )
    
    if 'Updated On' in result.columns and not pd.api.types.is_datetime64_dtype(result['Updated On']):
        result['Updated On'] = pd.to_datetime(result['Updated On'], errors='coerce')
    
    return result

# Rename
def rename_and_reorganize_columns(df):

    result = df.copy()
    
    # Rename The Columns
    rename_map = {
        'ID': 'id',
        'Case Number': 'case_number',
        'Date': 'date',
        'Block': 'block',
        'IUCR': 'iucr_code',
        'Primary Type': 'crime_type',
        'Description': 'crime_description',
        'Location Description': 'location_description',
        'Arrest': 'is_arrest',
        'Domestic': 'is_domestic',
        'Beat': 'police_beat',
        'District': 'police_district',
        'Ward': 'ward',
        'Community Area': 'community_area',
        'FBI Code': 'fbi_code',
        'X Coordinate': 'x_coordinate',
        'Y Coordinate': 'y_coordinate',
        'Year': 'year',
        'Updated On': 'last_updated',
        'Latitude': 'latitude',
        'Longitude': 'longitude',
        'Location': 'location'
    }
    
    cols_to_rename = {k: v for k, v in rename_map.items() if k in result.columns}
    result = result.rename(columns=cols_to_rename)
    
    # Core Features
    core_features = [
        'id', 'case_number', 'date', 'year', 'month', 'day', 'dayofweek',
        'hour', 'timeofday', 'season', 'is_weekend',
        'crime_type', 'crime_description', 'iucr_code', 'fbi_code',
        'location_description', 'is_arrest', 'is_domestic',
        'police_beat', 'police_district', 'ward', 'community_area',
        'latitude', 'longitude', 'x_coordinate', 'y_coordinate', 'location',
        'last_updated'
    ]
    
    cols_order = [col for col in core_features if col.lower() in [c.lower() for c in result.columns]]
    
    for col in result.columns:
        if col.lower() not in [c.lower() for c in cols_order]:
            cols_order.append(col)
    
    # Reorder
    result = result[[col for col in cols_order if col in result.columns]]
    
    return result

# Filter Time 
def filter_time_window(df, years=None):

    result = df.copy()
    
    if years is not None and 'year' in result.columns:
        result = result[result['year'].isin(years)]
        print(f"Filtered data to years: {years}")
        print(f"Remaining records: {len(result)}")
    
    return result

#----- Phase 2: Advanced Preprocessing -----#

## Outliers
def handle_outliers(df, cols=None):

    result = df.copy()
    
    if cols is None:
        cols = result.select_dtypes(include=['int', 'float']).columns
    else:
        cols = [col for col in cols if col in result.columns and 
                np.issubdtype(result[col].dtype, np.number)]
    
    outlier_report = {}
    
    for col in cols:

        if col in ['id', 'case_number', 'year', 'month', 'day', 'hour', 
                   'police_beat', 'police_district', 'ward', 'community_area']:
            continue
            
        # IQR Method
        Q1 = result[col].quantile(0.25)
        Q3 = result[col].quantile(0.75)
        IQR = Q3 - Q1
        
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        
        outliers = result[(result[col] < lower_bound) | (result[col] > upper_bound)]
        outliers_count = len(outliers)
        
        outlier_report[col] = {
            'count': outliers_count,
            'percentage': (outliers_count / len(result)) * 100,
            'lower_bound': lower_bound,
            'upper_bound': upper_bound
        }
        
        if outlier_report[col]['percentage'] > 10:
            continue
        
        result[col] = result[col].clip(lower=lower_bound, upper=upper_bound)
    
    return result, outlier_report

# Validate Geospatial Data
def validate_geospatial_data(df):
    result = df.copy()
    
    # Chicago Bounds
    chicago_bounds = {
        'lat_min': 41.6,
        'lat_max': 42.1,
        'lon_min': -87.9,
        'lon_max': -87.5
    }
    
    report = {
        'original_count': len(result),
        'invalid_coords': 0,
        'corrected_coords': 0
    }
    
    if 'latitude' in result.columns:
        invalid_lat = ((result['latitude'] < chicago_bounds['lat_min']) | 
                       (result['latitude'] > chicago_bounds['lat_max']))
        report['invalid_lat_count'] = invalid_lat.sum()

        if report['invalid_lat_count'] > 0:
            result.loc[result['latitude'] < chicago_bounds['lat_min'], 'latitude'] = chicago_bounds['lat_min']
            result.loc[result['latitude'] > chicago_bounds['lat_max'], 'latitude'] = chicago_bounds['lat_max']
            report['corrected_coords'] += report['invalid_lat_count']
    
    if 'longitude' in result.columns:
        invalid_lon = ((result['longitude'] < chicago_bounds['lon_min']) | 
                       (result['longitude'] > chicago_bounds['lon_max']))
        report['invalid_lon_count'] = invalid_lon.sum()
        
        if report['invalid_lon_count'] > 0:
            result.loc[result['longitude'] < chicago_bounds['lon_min'], 'longitude'] = chicago_bounds['lon_min']
            result.loc[result['longitude'] > chicago_bounds['lon_max'], 'longitude'] = chicago_bounds['lon_max']
            report['corrected_coords'] += report['invalid_lon_count']
    
    # Handle location column if it exists and needs updating
    if 'latitude' in result.columns and 'longitude' in result.columns and 'location' in result.columns:
        mask = ((result['latitude'] <= chicago_bounds['lat_min']) | 
                (result['latitude'] >= chicago_bounds['lat_max']) |
                (result['longitude'] <= chicago_bounds['lon_min']) | 
                (result['longitude'] >= chicago_bounds['lon_max']))
        
        if pd.api.types.is_categorical_dtype(result['location']):
            result['location'] = result['location'].astype(str)
        
        # Update locations
        new_locations = result.loc[mask].apply(
            lambda row: f"({row['latitude']}, {row['longitude']})", axis=1
        )
        result.loc[mask, 'location'] = new_locations
    
    return result, report

# Consolidate Categories
def consolidate_categories(df):

    result = df.copy()
    
    report = {}
    
    # Crime Type
    if 'crime_type' in result.columns:
        original_categories = result['crime_type'].nunique()
        
        crime_type_mapping = {
            'THEFT': 'THEFT',
            'MOTOR VEHICLE THEFT': 'THEFT',
            'BURGLARY': 'THEFT',
            'ROBBERY': 'THEFT',
            
            'BATTERY': 'VIOLENT CRIME',
            'ASSAULT': 'VIOLENT CRIME',
            'HOMICIDE': 'VIOLENT CRIME',
            'WEAPONS VIOLATION': 'VIOLENT CRIME',
            'CRIM SEXUAL ASSAULT': 'VIOLENT CRIME',

            'CRIMINAL DAMAGE': 'PROPERTY CRIME',
            'CRIMINAL TRESPASS': 'PROPERTY CRIME',
            'ARSON': 'PROPERTY CRIME',
            
            'NARCOTICS': 'DRUG RELATED',
            
            'DECEPTIVE PRACTICE': 'FRAUD',
            
            'OTHER OFFENSE': 'OTHER',
            'PUBLIC PEACE VIOLATION': 'OTHER',
            'INTERFERENCE WITH PUBLIC OFFICER': 'OTHER',
            'LIQUOR LAW VIOLATION': 'OTHER',
            'GAMBLING': 'OTHER',
            'PROSTITUTION': 'OTHER',
            'SEX OFFENSE': 'OTHER',
            'STALKING': 'OTHER',
        }
        

        lowercase_mapping = {k.lower(): v for k, v in crime_type_mapping.items()}
        result['crime_category'] = result['crime_type'].apply(
            lambda x: lowercase_mapping.get(x.lower(), 'OTHER') if isinstance(x, str) else 'OTHER'
        )
        
        # New Categories
        new_categories = result['crime_category'].nunique()
        
        report['crime_type'] = {
            'original_categories': original_categories,
            'new_categories': new_categories,
            'reduction_percentage': ((original_categories - new_categories) / original_categories) * 100
        }
    
    # Location
    if 'location_description' in result.columns:

        original_categories = result['location_description'].nunique()
        
        location_mapping = {
            'RESIDENCE': 'RESIDENTIAL',
            'APARTMENT': 'RESIDENTIAL',
            'HOUSE': 'RESIDENTIAL',
            'RESIDENCE-GARAGE': 'RESIDENTIAL',
            'RESIDENTIAL YARD': 'RESIDENTIAL',
            'PORCH': 'RESIDENTIAL',
            'DRIVEWAY': 'RESIDENTIAL',
            
            'STREET': 'STREET',
            'SIDEWALK': 'STREET',
            'ALLEY': 'STREET',
            'PARKING LOT': 'STREET',
            'PARKING LOT/GARAGE(NON.RESID.)': 'STREET',
            'CTA PLATFORM': 'STREET',
            'CTA STATION': 'STREET',
            'CTA BUS': 'STREET',
            'CTA TRAIN': 'STREET',
           
            'RESTAURANT': 'COMMERCIAL',
            'BAR OR TAVERN': 'COMMERCIAL',
            'RETAIL STORE': 'COMMERCIAL',
            'GROCERY STORE': 'COMMERCIAL',
            'DEPARTMENT STORE': 'COMMERCIAL',
            'GAS STATION': 'COMMERCIAL',
            'HOTEL/MOTEL': 'COMMERCIAL',
            'BANK': 'COMMERCIAL',
            'CONVENIENCE STORE': 'COMMERCIAL',
            'DRUG STORE': 'COMMERCIAL',
            
            'SCHOOL': 'INSTITUTIONAL',
            'COLLEGE/UNIVERSITY': 'INSTITUTIONAL',
            'HOSPITAL': 'INSTITUTIONAL',
            'NURSING HOME': 'INSTITUTIONAL',
            'CHURCH/SYNAGOGUE/PLACE OF WORSHIP': 'INSTITUTIONAL',
            'LIBRARY': 'INSTITUTIONAL',
            
            'PARK PROPERTY': 'PUBLIC SPACE',
            'ATHLETIC CLUB': 'PUBLIC SPACE',
            'FOREST PRESERVE': 'PUBLIC SPACE',
            'GOVERNMENT BUILDING': 'PUBLIC SPACE',
            'POLICE FACILITY / VEHICLE PARKING LOT': 'PUBLIC SPACE',
           
            'VEHICLE-COMMERCIAL': 'VEHICLE',
            'VEHICLE-DELIVERY TRUCK': 'VEHICLE',
            'VEHICLE NON-COMMERCIAL': 'VEHICLE'
        }
        

        lowercase_mapping = {k.lower(): v for k, v in location_mapping.items()}
        result['location_category'] = result['location_description'].apply(
            lambda x: lowercase_mapping.get(x.lower() if isinstance(x, str) else '', 'OTHER')
        )
        

        new_categories = result['location_category'].nunique()
        
        report['location_description'] = {
            'original_categories': original_categories,
            'new_categories': new_categories,
            'reduction_percentage': ((original_categories - new_categories) / original_categories) * 100
        }
    
    return result, report

## Data Types
def optimize_data_types(df):

    result = df.copy()
    
    for col in result.select_dtypes(include=['object']).columns:
        if col not in ['id', 'case_number', 'location']:  
            if result[col].nunique() < 0.5 * len(result): 
                result[col] = result[col].astype('category')

    for col in ['is_arrest', 'is_domestic', 'is_weekend']:
        if col in result.columns:
            result[col] = result[col].astype('bool')
    
    # Integer
    int_cols = ['year', 'month', 'day', 'hour', 'police_beat', 'police_district', 'ward', 'community_area']
    for col in [c for c in int_cols if c in result.columns]:

        c_min = result[col].min()
        c_max = result[col].max()
        
        if c_min >= 0:
            if c_max < 255:
                result[col] = result[col].astype(np.uint8)
            elif c_max < 65535:
                result[col] = result[col].astype(np.uint16)
        else:
            if c_min > -128 and c_max < 127:
                result[col] = result[col].astype(np.int8)
            elif c_min > -32768 and c_max < 32767:
                result[col] = result[col].astype(np.int16)
    
    float_cols = ['latitude', 'longitude', 'x_coordinate', 'y_coordinate']
    for col in [c for c in float_cols if c in result.columns]:
        result[col] = result[col].astype(np.float32)
    
    return result
```

```python
#----- Phase 3: Save Processed Data -----#

def save_processed_data(df, filename, info=None):
 
    filepath = os.path.join(PROCESSED_DIR, filename)
    
    # Save Data
    if filename.endswith('.csv'):
        df.to_csv(filepath, index=False)
    elif filename.endswith('.parquet'):
        df.to_parquet(filepath, index=False)
    elif filename.endswith('.pkl'):
        with open(filepath, 'wb') as f:
            pickle.dump(df, f)
    else:
        with open(filepath + '.pkl', 'wb') as f:
            pickle.dump(df, f)

    if info:
        info_filepath = filepath.rsplit('.', 1)[0] + '_info.json'
        pd.Series(info).to_json(info_filepath)
```

## Phase 1: Fundamental Cleaning

```python
# Missing Values
df_clean, missing_report = handle_missing_values(df)
print(f"Missing values before: {sum(missing_report['missing_before'])}")
print(f"Missing values after: {sum(missing_report['missing_after'])}")

# Date
df_clean = standardize_datetime(df_clean)
print(f"Added temporal features: Year, Month, Day, DayOfWeek, Hour, TimeOfDay, IsWeekend, Season")

# Rename
df_clean = rename_and_reorganize_columns(df_clean)
print(f"Columns renamed to more intuitive names and reordered")

# Filter Time
df_clean = filter_time_window(df_clean, years=[2012, 2013, 2014, 2015, 2016, 2017])

# Save Phase 1
save_processed_data(df_clean, 'chicago_crimes_phase1_cleaned.parquet', {
    'date_processed': datetime.now().strftime('%Y-%m-%d %H:%M:%S'),
    'data_shape': df_clean.shape,
    'processing_stage': 'Phase 1 - Fundamental Cleaning'
})
```

## Phase 2: Advanced Preprocessing

```python
# Outliers
df_cleaned, outlier_report = handle_outliers(df_clean)
print(f"Processed outliers in {len(outlier_report)} columns")
for col, report in outlier_report.items():
    if report['percentage'] > 0:
        print(f"  {col}: {report['count']} outliers ({report['percentage']:.2f}%)")

# Geospatial Data
df_cleaned, geo_report = validate_geospatial_data(df_cleaned)
print(f"Invalid coordinates: {geo_report.get('invalid_lat_count', 0) + geo_report.get('invalid_lon_count', 0)}")
print(f"Corrected coordinates: {geo_report['corrected_coords']}")

# Merge Categories
df_cleaned, category_report = consolidate_categories(df_cleaned)
for field, report in category_report.items():
    print(f"  {field}: Reduced from {report['original_categories']} to {report['new_categories']} categories")
    print(f"  Reduction: {report['reduction_percentage']:.2f}%")

# Data Types
df_cleaned = optimize_data_types(df_cleaned)
memory_usage = df_cleaned.memory_usage(deep=True).sum() / (1024 * 1024)
print(f"Memory usage after optimization: {memory_usage:.2f} MB")

# Save Phase 2
save_processed_data(df_cleaned, 'chicago_crimes_cleaned.parquet', {
    'date_processed': datetime.now().strftime('%Y-%m-%d %H:%M:%S'),
    'data_shape': df_cleaned.shape,
    'processing_stage': 'Phase 2 - Advanced Preprocessing',
    'memory_usage_mb': memory_usage
})
```
